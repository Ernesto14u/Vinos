```python
# Función para ponerle la caliad al vino, mediante sus puntos, en una nueva columna
def nuevafun(a):
   if a>=95 :
       return(" Excepcional")
   elif(a>=90) and (a< 95):
       return(" Excelente")
   elif(a>=80) and (a< 90):
       return(" Muy Bueno")
   else:
       return("No Compiten")


vino['rank']= vino['points'].apply(nuevafun)
vino


```

## US

```python
vinos_us = vinosII.loc[vinosII['country']== 'US']
```

```python
vinos_us
```

```python
vinos_us[['title','points','price']]
```

```python
sns.scatterplot(data=vinos_us, x="price", y="points");
```

```python
vinos_us.corr('pearson')
```

```python
# Prueba con US
#Nos quedamos sólo con la columna description

descriptions_us=vinos_us.description

# Unimos toda la columna en un texto
descriptions_text_us = vinos_us['description'].str.lower().str.cat(sep=' ')

#Convertimos el texto en una lista de palabras separadas

descriptions_list_us=descriptions_text_us.split()

#Eliminamos las palabras de conexión y palabras comunes

descriptions_list_us_nosw= filter(lambda w: not w in stopwords_total,descriptions_list_us)
descriptions_list_us_nosw=list(descriptions_list_us_nosw)


#10 palabras más frecuentes 
from collections import Counter
Counter = Counter(descriptions_list_us_nosw) 
most_occur_us = Counter.most_common(10)

#Convertir lista en Data frame

most_occur_df_us = pd.DataFrame(most_occur_us,columns=['Word', 'Frequency'])
most_occur_df_us.head()
```
